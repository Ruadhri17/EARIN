```python
import os

import gym

from stable_baselines3 import PPO, A2C
from stable_baselines3.common.env_util  import make_vec_env
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.callbacks import EvalCallback
from stable_baselines3.common.evaluation import evaluate_policy
```

```python
saved_models_path = os.path.join('Training_PPO', 'Saved Models')
logs_path = os.path.join('Training_PPO', 'Logs')
last_model_path = os.path.join('Training_PPO', 'Saved Models', 'last_model')
```

```python
env = gym.make('CarRacing-v0')
env = DummyVecEnv([lambda: env])
```

```python
eval_callback = EvalCallback(env, 
                            best_model_save_path=saved_models_path, 
                            log_path=logs_path, 
                            n_eval_episodes=5, 
                            eval_freq=10000, 
                            verbose=1, 
                            deterministic=True, 
                            render=False)

model = PPO('CnnPolicy', env, verbose=1, tensorboard_log=logs_path)
model.learn(500000, callback=eval_callback)
model.save(last_model_path)
del model
```

```python
model_path = os.path.join('Training_PPO', 'Saved Models', 'best_model')
model = PPO.load(model_path, env=env)
```

```python
evalue = evaluate_policy(model, env, n_eval_episodes=10, render=True)
env.close()
evalue
```

```python

obs = env.reset()
done = False
score = 0;    
while not done:
    env.render()
    action, _ = model.predict(obs)
    obs, reward, done, info = env.step(action)
    score+= reward
print("Final Score: {}".format(score))
env.close()
```

```python
saved_models_path = os.path.join('Training_A2C', 'Saved Models')
logs_path = os.path.join('Training_A2C', 'Logs')
last_model_path = os.path.join('Training_A2C', 'Saved Models', 'last_model')
```

```python
env = make_vec_env('CarRacing-v0', n_envs=4)
```

```python

eval_callback = EvalCallback(env, 
                            best_model_save_path=saved_models_path, 
                            log_path=logs_path, 
                            n_eval_episodes=5, 
                            eval_freq=10000, 
                            verbose=1, 
                            deterministic=True, 
                            render=False)
model = A2C('CnnPolicy', env, verbose=1, tensorboard_log=logs_path)
model.learn(total_timesteps=500000, callback=eval_callback)
model.save(last_model_path)
del model
```

```python
model_path = os.path.join('Training_A2C', 'Saved Models', 'best_model')
model = A2C.load(model_path, env=env)
```

```python
evalue = evaluate_policy(model, env, n_eval_episodes=10, render=True)
env.close()
evalue
```

```python
obs = env.reset()
done = False
score = 0;    
while True:
    env.render()
    action, _ = model.predict(obs)
    obs, reward, done, info = env.step(action)
    score+= reward
    if not done.all():
        break;
print("Final Score: {}".format(score))
env.close()
```
